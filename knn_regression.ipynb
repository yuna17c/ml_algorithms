```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from knn_regression import k_smallest, knn_regression, run_model, plot_data
```

```python
# Load datasets
X_test_D = np.loadtxt(open("datasets/X_test_D.csv", "rb"), delimiter=",")
X_test_E = np.loadtxt(open("datasets/X_test_E.csv", "rb"), delimiter=",")
X_train_D = np.loadtxt(open("datasets/X_train_D.csv", "rb"), delimiter=",")
X_train_E = np.loadtxt(open("datasets/X_train_E.csv", "rb"), delimiter=",")
Y_test_D = np.loadtxt(open("datasets/Y_test_D.csv", "rb"), delimiter=",")
Y_test_E = np.loadtxt(open("datasets/Y_test_E.csv", "rb"), delimiter=",")
Y_train_D = np.loadtxt(open("datasets/Y_train_D.csv", "rb"), delimiter=",")
Y_train_E = np.loadtxt(open("datasets/Y_train_E.csv", "rb"), delimiter=",")
X_test_F = np.loadtxt(open("datasets/X_test_F.csv", "rb"), delimiter=",")
X_train_F = np.loadtxt(open("datasets/X_train_F.csv", "rb"), delimiter=",")
Y_test_F = np.loadtxt(open("datasets/Y_test_F.csv", "rb"), delimiter=",")
Y_train_F = np.loadtxt(open("datasets/Y_train_F.csv", "rb"), delimiter=",")
```

```python
# dataset D
run_model(X_train_D, Y_train_D, X_test_D, Y_test_D)
```

From the graphs, we can deduce that the data is linear, so the least squares solution always performs better than KNN regression solutions. Least squares linear regression minimizes the residual for a linear model, and therefore it will outperform the KNN regression solution. 

```python
# dataset E
run_model(X_train_E, Y_train_E, X_test_E, Y_test_E)
```

KNN regression solution performs very well with most  values. For all 1    9, KNN regression 
outperforms least-squares linear regression solution. This is because the data is not linear. Since least-squares linear regression assumes a linear relation, it does not perform very well for non-linear data.

```python
# dataset F
run_model(X_train_F, Y_train_F, X_test_F, Y_test_F, False)
```

In this case, neither approach performs well. As the dimensionality increases, the distance becomes 
more spread out, and the data points become more equidistant from each other. This works against the 
KNN algorithm as finding the nearest neighbours is less meaningful when all points are near each other. Least-squares approach does not perform well as the data is not linear.
